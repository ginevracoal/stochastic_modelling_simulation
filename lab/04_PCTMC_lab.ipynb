```python
def lotkavolterra():
    lv = Model()
    # Adding variables
    lv.add_variable("prey", 6)
    lv.add_variable("predator", 4)
    
    # Adding parameters (rates)
    lv.add_parameter("aggress", 6)
    lv.add_parameter("birth", 3)
    
    # Adding transitions, using a dictionary to represent the update vector
    
    # transition: (1,1) -> (0,1) , update vector = (-1, 0)
    # qua sto sbagliando!!
    lv.add_transition({"prey":-1,"predator":0}, "aggress*predator")
    lv.add_transition({"prey":0,"predator":1}, "birth*prey")
    
    # adding observables
    lv.add_observable("alive animals","prey"+"predator")

    return lv.finalize_initialization()          
```
